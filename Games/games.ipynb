## Предобработка данных

#### Изменение названий столбцов

```python
# приведение названий столбцов к нижнему регистру
data.columns = data.columns.str.lower()
```

#### Обработка пропусков

```python
data.isna().sum()  # подсчет количества пропусков в столбцах
```

```python
data = data.query('~year_of_release.isna()')
# удалим строки, в которых пропущен год релиза игры, пропусков немного
# они не представляют пользы при анализе
data['year_of_release'].isna().sum() 
```

```python
data[data['name'].isna()]  # посмотрим строки с пропусками в столбце name
```

Заметим, что пропуски в столбце с именем и жанром в одних и тех же строках. Можно их удалить

```python
data = data[~data['name'].isna()]
```

```python
data.query('rating.isna()')
```

Пропуски в столбце с возрастной категорией не получится заменить на какое-то определенное значение. Удалять строки с ними тоже неправильно, потому что потеряется много нужных данных. Заменим пропущенные значения на *unspecified*

```python
data['rating'] = data['rating'].fillna('unspecified')
```

Пропуски в столбцах с оценками можно оставить, потому что это означает, что либо их нет, либо слишком мало. Замена на что-то может помешать в дальнейших исследованиях оценок игр. Значение *tbd* в столбце user_score может означать, что оценок пока нет (потому что игра новая), либо оценок совсем нет или их мало, чтобы сформировать рейтинг. Заменим его на Nan

```python
data['user_score'] = data['user_score'].replace('tbd', None)
```

#### Изменение типа данных столбцов

```python
data['year_of_release'] = data['year_of_release'].astype(dtype='Int64')
```

```python
data['critic_score'] = data['critic_score'].astype(dtype='Int64')
```

```python
data['user_score'] = pd.to_numeric(data['user_score'], errors='ignore')
```

#### Обработка дубликатов

```python
data = data.drop_duplicates()  # удаление явных дубликатов
```

Рассмотрим неявные дубликаты

```python
for i in data.loc[data['name'].duplicated()]['name'].unique():
    a = data.loc[data['name'] == i]
    if len(a.loc[a['platform'].duplicated()]):
        print(a)
```

Неявных дубликатов нет

#### Суммарные продажи во всех регионах

```python
data['sum_sales'] = data.loc[:, 'na_sales':'other_sales'].sum(axis=1)
# создание столбца с общим количеством продаж по всем регионам
```

#### Природа выбросов в данных

Проверим сильно выделяющиеся значения в столбце с суммарными продажами во всех регионах

```python
data.loc[data['sum_sales'] > 10]
```

Выделяющиеся значения - успешные игры, которые действительно продавались в огромных количествах. Их немного, поэтому медианное значение сильно отличается от максимальных

## Исследовательский анализ

#### Количество игр в разные годы

```python
data[~data['name'].duplicated()]\
    .groupby('year_of_release')\
    .count()['name']\
    .plot(kind='bar', figsize=(13, 6))
plt.xlabel('Год релиза', fontsize=10)
plt.ylabel('Количество выпущенных игр', fontsize=10)
plt.title('Количество выпущенных игр по годам', fontsize=15)
plt.show()
```

По графику видно, что до 1994 года компьютерных игр было очень мало. Это связано с отсутствием технологий для создания игр, а также с тем, что компьютеры или консоли были далеко не у всех. Пик приходится на 2009-2010 года, однако после идет резкий спад. Причиной этому может быть переход на другие платформы (например, смартфоны), потеря интереса к играм или увеличение срока создания игр (из-за использования более сложных технологий).

#### Продажи игр на разных платформах

Для планирования кампании на 2017 год данные за все года не понадобятся, так как даже в последние 5 лет были заметные изменения в индустрии. Достаточно взять данные за последние 3 года

```python
data.groupby('platform')['sum_sales'].sum().sort_values().plot(kind='barh', figsize=(13, 6))
plt.xlabel('Количество проданных копий', fontsize=10)
plt.ylabel('Платформы', fontsize=10)
plt.title('Количество проданных копий игр на разных платформах за все время', fontsize=15)
plt.show()
```

```python
top_data = data.loc[data['platform']\
                    .isin(data.groupby('platform')['sum_sales']\
                    .sum().sort_values().index[-5:])]
# таблица 5 платформ с наибольшими суммарными продажами
```

```python
platform_data = top_data.pivot_table(index='year_of_release',
                                     columns='platform',
                                     values='sum_sales')
platform_data.plot(kind='bar', figsize=(13, 6))
plt.xlabel('Год релиза', fontsize=10)
plt.ylabel('Количество проданных копий', fontsize=10)
plt.title('Количество проданных копий игр для топ-5 платформ по годам', fontsize=15)
plt.show()
```

По графику видно, что в среднем платформа актуальна около 10 лет. После появляются новые, улучшенные платформы, забирающие рынок.

#### Зависимость количества проданных копий от оценок для самой популярной платформы

Посмотрим, как зависят продажи игр на платформе *PS4* от оценок пользователей и критиков

```python
actual_data_ps4.plot(kind='scatter',
                     y='sum_sales',
                     x='critic_score',
                     alpha=0.6,
                     figsize=(13, 6))
plt.ylabel('Количество проданных копий', fontsize=10)
plt.xlabel('Оценки критиков', fontsize=10)
plt.title('Зависимость продаж от оценок критиков', fontsize=15)
plt.show()
```

По диаграммам видно, что между продажами и оценками критиков есть прямая зависимость. Однако оценки пользователей не всегда определяют успешность продаж игр

#### Зависимость количества проданных копий от оценок для самой остальных платформ

```python
platforms = actual_data.pivot_table(index='platform', values='sum_sales')\
                        .sort_values(by='sum_sales', ascending=False).index[1:6]
for platform in platforms:
    curr_data = actual_data.query('platform == platform')
    fig, ax = plt.subplots(1, 2, figsize=(20, 5))
    ax[0].scatter(y=curr_data['sum_sales'],
                  x=curr_data['user_score'],
                  alpha=0.6)
    ax[0].set_title(f'Зависимость продаж платформы {platform} от оценки пользователей', fontsize=13)
    ax[0].set_xlabel('Оценки пользователей', fontsize=12)
    ax[0].set_ylabel('Количество проданных копий', fontsize=12)

    curr_data_cr = curr_data.loc[~curr_data['critic_score'].isna()]
    ax[1].scatter(y=curr_data_cr['sum_sales'],
                  x=curr_data_cr['critic_score'],
                  alpha=0.6)
    ax[1].set_title(f'Зависимость продаж платформы {platform} от оценки критиков', fontsize=13)
    ax[1].set_xlabel('Оценки критиков', fontsize=12)
    ax[1].set_ylabel('Количество проданных копий', fontsize=12)

    plt.show()

    print(curr_data[['sum_sales', 'critic_score', 'user_score']].corr())
```

Аналогично с платформой *PS4* продажи прямо зависят от оценки критиков, а отзывы пользователей не влияют. Но на *3DS* и *WiiU* корреляция между оценками пользователей и продажами положительна

Больше всего игр в жанре *Action*, *Puzzle* самый непопулярный жанр. По продажам также лидирует жанр *Action*, однако *Shooter* не сильно отстает, несмотря на большое различие в количестве выпущенных игр.

Самым прибыльным жанром является *Shooter*, а самым невыгодным *Adventure*. *Action* же больше всего продается только благодаря количеству игр.

## Портрет пользователей разных регионов

#### Самые популярные платформы в регионах

```python
regions_sales = ['na_sales', 'eu_sales', 'jp_sales']
regions = ['Северная Америка', 'Европа', 'Япония']
```

```python
fig, ax = plt.subplots(1, 3, figsize=(25, 10))
fig.suptitle('Самые популярные платформы по регионам', fontsize=30)

for i, region in enumerate(regions_sales):
    region_data = actual_data.pivot_table(index='platform',
                                          values=region,
                                          aggfunc='sum').sort_values(by=region, ascending=False)
    ax[i].pie(region_data[:5][region],
              autopct='%.2f%%',
              labels=region_data[:5].index,
              wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
              textprops={'size': 'x-large'})
    ax[i].set_title(regions[i], fontsize=20)

plt.show()
```

Самыми популярными платформами являются: Северная Америка - PS4, Европа - PS4, Япония - 3DS

## Гипотезы

#### Первая гипотеза

**H0 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**

**H1 - Средние пользовательские рейтинги платформ Xbox One и PC отличаются**

```python
samp1 = actual_data.loc[actual_data['platform'] == 'PC']['user_score'].dropna()
samp2 = actual_data.loc[actual_data['platform'] == 'XOne']['user_score'].dropna()

alpha = .05  # стандартный критический уровень статистической значимости

results = st.ttest_ind(samp1, samp2)
print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем гипотезу')
else:
    print('Не получилось отвергнуть гипотезу')
```

Нулевую гипотезу не получилось отвергнуть, значит нельзя утверждать, что пользовательские рейтинги платформ *XOne* и *PC* сильно отличаются 

#### Вторая гипотеза

**H0 - Средние пользовательские рейтинги жанров Action и Sports одинаковы**

**H1 - Средние пользовательские рейтинги жанров Action и Sports отличаются**

```python
samp1 = actual_data.loc[actual_data['genre'] == 'Action']['user_score'].dropna()
samp2 = actual_data.loc[actual_data['genre'] == 'Sports']['user_score'].dropna()

alpha = .05  # стандартный критический уровень статистической значимости

results = st.ttest_ind(samp1, samp2)
print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем гипотезу')
else:
    print('Не получилось отвергнуть гипотезу')
```

Нулевая гипотеза не подтвердилась. Вероятность получить одинаковые значения слишком мала. 

Нулевые гипотезы формулировались, чтобы определять равенство двух совокупностей. В качестве альтернативной гипотезы использовалось отрицание равенства (т.е значения либо больше, либо меньше). Критический уровень значимости был взять стандартный - 5%. Далее сравнивался полученный уровень значимости для двух выборок и сравнивался с критическим. Если вероятность получить равные значения достаточно большая, то гипотеза не отвергается, иначе - отвергается
