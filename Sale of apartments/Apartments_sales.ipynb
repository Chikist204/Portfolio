В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

### Предобработка данных

#### Изменение названий и типа данных столбцов 

```python
data.isna().sum()
```

```python
data = data.rename(columns={'parks_around3000': 'parks_around', 
             'ponds_around3000': 'ponds_around', 
             'cityCenters_nearest': 'city_centers_nearest'})
data.head() # изменение некорректных названий столбцов
```

#### Обработка аномалий и пропусков

##### ***ceiling_height***

```python
sorted(data['ceiling_height'].unique())
```

```python
data.query('2 < ceiling_height < 8.3')['ceiling_height'].mean() # среднее значение высоты потолков без аномальных значений
```

```python
data.loc[data['ceiling_height'].isin(range(20, 100)), 'ceiling_height'] = data['ceiling_height'] / 10
# возможная причина слишком больших значений - вместо вещественных значений были введены целые
data.loc[data['ceiling_height'] < 2, 'ceiling_height'] = data['ceiling_height'].median()
data.loc[data['ceiling_height'] > 8, 'ceiling_height'] = data['ceiling_height'].median()
# заменим остальные аномалии медианным значением, чтобы не сильно повлиять на среднее значение и не потерять нужные данные
```

```python
data['ceiling_height'].mean() # посчитаем среднее значения без аномальных значений
```

```python
data['ceiling_height'] = data['ceiling_height'].fillna(2.73) # заполним пропуски средним значением
```

```python
data['ceiling_height'].isna().sum() # проверим значения в столбце еще раз
```

В столбце с высотой потолков было много аномальных значений. Наличие некоторых можно объяснить забытой запятой перед дробной частью. Значения были исправлены. Однако не все значения можно объяснить поэтому было решено заменить их на медианные значения, чтобы не потерять другие данные

```python
data[data['floors_total'].isna()] # нельзя ни с чем связать пропуски
```

```python
data = data[~data['floors_total'].isna()]
len(data[data['floors_total'].isna()]) # проверка наличия пропусков
```

Посмотрим на значения в стобце *floors_total*

```python
data['floors_total'].unique()
```

```python
data['floors_total'].dtypes
```

Все значения целые, но тип данных *float*. Изменим тип данных на *int*. Это позволит немного ускорить операции с данными, а также избежать возможных ошибок в вычислении, связанных с особенностью хранения данных с плавающей точкой

```python
data['floors_total'] = data['floors_total'].astype("int", errors='ignore')
data['floors_total'].dtype
```

##### ***living_area***

```python
data['living_area'].describe()
```

```python
data.loc[data['living_area'] < 5, 'living_area'] = None # заполним аномально маленькие значения None
```

Попробуем заполнить пропуски рассчетом по известным значениям

```python
for index in data.loc[(data['living_area'].isna()) & ~(data['kitchen_area'].isna())].index:
    data.loc[index, 'living_area'] = data.loc[index, 'total_area'] - data.loc[index, 'kitchen_area']
# рассчет площади жилой части в строках, где пропущено значение
```

```python
data['living_area'].isna().sum() # проверим оставшееся количество пропусков
```

##### ***kitchen_area***

По стандартам размеры кухни не могут быть меньше 5 м^2. Заменим такие данные на *None*

```python
data['kitchen_area'].describe()
```

Попробуем заполнить пропуски рассчетом по известным значениям

```python
for index in data.loc[(data['kitchen_area'].isna()) & ~(data['living_area'].isna())].index:
    data.loc[index, 'kitchen_area'] = data.loc[index, 'total_area'] - data.loc[index, 'living_area']
```

По стандартам площадь кухни не может быть меньше 5 м2. Заменим аномальные значения None

```python
data.loc[data['kitchen_area'] < 5, 'kitchen_area'] = None
```

```python
data['kitchen_area'].isna().sum() # проверим оставшееся количество пропусков
```

##### ***airports_nearest***

```python
data['airports_nearest'].describe()
```

Заменим пропуски на нули

```python
data.loc[data['airports_nearest'].isna(), 'airports_nearest'] = 0
```

```python
len(data['locality_name'].unique())
```

Заполним нулевые значения на средние значения для районов, в которых расположены помещения

```python
distance_airport = data.pivot_table(index='locality_name', values='airports_nearest', aggfunc='mean')
for index in data[data['airports_nearest'] == 0].dropna().index:
    data.loc[index, 'airports_nearest'] = distance_airport.loc[data.loc[index, 'locality_name'], 'airports_nearest']
```

Заметим, что некоторые названия дублируются из-за использования букв "е" и "ё". Заменим все на е и избавимся от дубликатов

```python
data['month_exposition'] = data['first_day_exposition'].dt.month_name()
data['month_exposition']
```

#### **Год публикации объявления**
